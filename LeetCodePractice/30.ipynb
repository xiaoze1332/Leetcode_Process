```python
class Solution:
    def findSubstring(self, s: str, words: List[str]) -> List[int]:
        # using the sliding window

        # judge if the words is empty
        if not words:
            return []
        
        # get the length of the words
        word_len = len(words[0])

        # get the length of the words list
        words_len = len(words)

        # get the length of the string
        s_len = len(s)

        # get the length of the substring
        sub_len = word_len * words_len

        # judge if the substring is longer than the string
        if sub_len > s_len:
            return []
        
        # get the words dict
        words_dict = {}
        for word in words:
            if word in words_dict:
                words_dict[word] += 1
            else:
                words_dict[word] = 1

        # get the result list
        result = []

        # get the substring
        for i in range(word_len):
            # get the left and right index
            left = i
            right = i

            # get the current dict
            cur_dict = {}

            # get the count
            count = 0

            while right + word_len <= s_len:
                word = s[right : right + word_len]
                right += word_len

                # judge if the word is in the words
                if word not in words_dict:
                    # clear the current dict
                    cur_dict = {}
                    count = 0
                    left = right
                else:
                    if word in cur_dict:
                        cur_dict[word] += 1
                    else:
                        cur_dict[word] = 1
                    count += 1

                    # judge if the count is greater than the words length
                    while cur_dict[word] > words_dict[word]:
                        cur_dict[s[left : left + word_len]] -= 1
                        count -= 1
                        left += word_len

                    # judge if the count is equal to the words length
                    if count == words_len:
                        result.append(left)

        return result
```
