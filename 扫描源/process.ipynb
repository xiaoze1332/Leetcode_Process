```python
from multiprocessing.pool import ThreadPool as Pool
import re
```

```python
txt_file = './input.txt'
```

```python
# 判断是否包含CCTV
def is_contains_CCTV(strs):
    if 'CCTV' in strs:
        return True
    return False
```

```python
# 判断是否包含国家名称，我们需要去掉带有国家名称的数据
def is_contains_country(strs):
    match_obj = r'\[.*\]'
    match_result = re.match(match_obj, strs)
    if match_result:
        return True
    return False
```

```python
# 判断是否包含中文。
# 选取带有中文的数据，并且去掉带有国家名称的数据
def is_contains_chinese(strs):
    if is_contains_country(strs):
        return False
    for _char in strs:
        if '\u4e00' <= _char <= '\u9fff':
            return True
    return False
```

```python
def process_func(line):
    if is_contains_chinese(line) or is_contains_CCTV(line):
        return line

    
```

```python
file = []

with open (txt_file, 'r', encoding='utf-8') as f:
    lines = f.readlines()

    # 使用多进程
    with Pool(2) as p:
        file = p.map(process_func, lines)


file = [x for x in file if x is not None]
file
```

```python
# 按照首字母排序
file.sort()
file
```

```python
# 输出到txt文件
with open('./output.txt', 'w', encoding='utf-8') as f:
    for line in file:
        f.write(line)
```
